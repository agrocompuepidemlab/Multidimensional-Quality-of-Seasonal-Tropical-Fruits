```python
fig, axs = plt.subplots(ncols=3,nrows=4, figsize=(20,25))
sns.set(font_scale=2)
plt.subplots_adjust(hspace=0.4)
plt.tight_layout(pad=2, w_pad=2, h_pad=2.5)
fig.subplots_adjust(top=0.94)


sns.boxplot(x="Fruit", y="PESO (g)", data=mamoncillo, ax=axs[0,0], color='#606639')
sns.boxplot(x="Fruit", y="VOLUMEN (cm"r'$^{-3}$'")", data=mamoncillo, ax=axs[0,1], color='#606639')
sns.boxplot(x="Fruit", y="DENSIDAD (g cm-3)", data=mamoncillo, ax=axs[0,2], color='#606639')
sns.boxplot(x="Fruit", y="D. POLAR (cm)", data=mamoncillo, ax=axs[1,0], color='#606639')
sns.boxplot(x="Fruit", y="D. ECUATORIAL (cm)", data=mamoncillo, ax=axs[1,1], color='#606639')
sns.boxplot(x="Fruit", y="GROSOR CÁSCARA (mm)", data=mamoncillo, ax=axs[1,2], color='#606639')
sns.boxplot(x="Fruit", y="PESO EPIDERMIS (g)", data=mamoncillo, ax=axs[2,0], color='#606639')
sns.boxplot(x="Fruit", y="PESO PULPA (g)", data=mamoncillo, ax=axs[2,1], color='#606639')
sns.boxplot(x="Fruit", y="PESO SEMILLAS (g)", data=mamoncillo, ax=axs[2,2], color='#606639')
sns.boxplot(x="Fruit", y="RELACIÓN EPI/PUL", data=mamoncillo, ax=axs[3,0], color='#606639')
sns.boxplot(x="Fruit", y="RELACIÓN PUL/SEM", data=mamoncillo, ax=axs[3,1], color='#606639')
sns.boxplot(x="Fruit", y="FIRMEZA (N)", data=mamoncillo, ax=axs[3,2], color='#606639')


axs[0,0].set_ylabel("Weight (g)", fontsize=20)
axs[0,1].set_ylabel("Volume (cm"r'$^{-3}$'")", fontsize=20)
axs[0,2].set_ylabel("Density (g cm"r'$^{-3}$'")", fontsize=20)
axs[1,0].set_ylabel("Polar diameter (cm)", fontsize=20)
axs[1,1].set_ylabel("Equatorial diameter (cm)", fontsize=20)
axs[1,2].set_ylabel("Shell thinning (mm)", fontsize=20)
axs[2,0].set_ylabel("Epidermis weight (g)", fontsize=20)
axs[2,1].set_ylabel("Pulp weight (g)", fontsize=20)
axs[2,2].set_ylabel("Seeds weight (g)", fontsize=20)
axs[3,0].set_ylabel("Epidermis/pulp ratio", fontsize=20)
axs[3,1].set_ylabel("Pulp/seeds ratio", fontsize=20)
axs[3,2].set_ylabel("Firmness (N)", fontsize=20)

axs[0,0].set(xlabel=None)
axs[0,1].set(xlabel=None)
axs[0,2].set(xlabel=None)
axs[1,0].set(xlabel=None)
axs[1,1].set(xlabel=None)
axs[1,2].set(xlabel=None)
axs[2,0].set(xlabel=None)
axs[2,1].set(xlabel=None)
axs[2,2].set(xlabel=None)
axs[3,0].set(xlabel=None)
axs[3,1].set(xlabel=None)
axs[3,2].set(xlabel=None)

plt.suptitle("Physic charactheristics", fontsize=35)

plt.legend()
plt.show()
```
